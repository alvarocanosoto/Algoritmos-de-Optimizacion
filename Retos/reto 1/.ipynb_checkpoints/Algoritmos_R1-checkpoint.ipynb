## Torres de Hanoi con Divide y vencerás

Resuelve el problema de las [Torres de Hanoi](https://es.wikipedia.org/wiki/Torres_de_Han%C3%B3i) para un número arbitrario de fichas utilizando la técnica de divide y vencerás. 

```python
# Torres de Hanoi
def hanoi(n, origen, destino, auxiliar):
    if n == 1:
        print(f"Mover ficha desde la {origen} a {destino}")
    else:
        hanoi(n-1, origen, auxiliar, destino)
        print(f"Mover ficha desde la {origen} a {destino}")
        hanoi(n-1, auxiliar, destino, origen)
```

```python
# Test
n_fichas = 4
hanoi(n_fichas, "Pila A", "Pila C", "Pila B")
```

## Sucesión de Fibonacci

Cálcula el n-ésimo término de la [Sucesión de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci) mediante un algoritmo recursivo y otro iterativo. Representa gráficamente cómo crece el tiempo de cómputo en función del número de términos para ambos algoritmos. 

```python
# imports
import time
import matplotlib.pyplot as plt
```

```python
#Sucesión_de_Fibonacci

def fibonacci_recursivo(n):
    if n <= 1:
        return n
    return fibonacci_recursivo(n - 1) + fibonacci_recursivo(n - 2)

def fibonacci_iterativo(n):
    if n <= 1:
        return n
    f0, f1 = 0, 1
    for _ in range(2, n + 1):
        f0, f1 = f1, f0 + f1
    return f1
```

```python
# Test
print(fibonacci_recursivo(15))
print(fibonacci_iterativo(15))
```

```python
# Tiempo de cómputo para ambos algoritmos
n_valores = list(range(1, 31))
tiempos_recursivo = []
tiempos_iterativo = []

for n in n_valores:
    # Tiempos para el algoritmo recursivo
    t_inicio = time.time()
    fibonacci_recursivo(n)
    t_fin = time.time()
    tiempos_recursivo.append(t_fin - t_inicio)

    # Tiempos para el algoritmo iterativo
    t_inicio = time.time()
    fibonacci_iterativo(n)
    t_fin = time.time()
    tiempos_iterativo.append(t_fin - t_inicio)

# Mostrar gráfica
plt.figure(figsize=(10, 6))
plt.plot(n_valores, tiempos_recursivo, label='Recursivo', marker='o')
plt.plot(n_valores, tiempos_iterativo, label='Iterativo', marker='o')
plt.xlabel('Número de términos de la Sucesión de Fibonacci (n)')
plt.ylabel('Tiempo de cómputo (segundos)')
plt.title('Comparación del tiempo de cómputo: Algoritmo Recursivo vs Iterativo')
plt.legend()
plt.grid(True)
plt.show()
```

## N-Reinas por técnica de vuelta atrás
Resuelve el [Problema de las N-Reinas](https://es.wikipedia.org/wiki/Problema_de_las_ocho_reinas) en un tablero de dimensión N mediante la técnica de la vuelta atrás (backtraking).

```python
# Problema de las N-Reinas
def resolver_n_reinas(tablero, fila, soluciones):
        # Comprobar si ya están todas las reinas colocadas
        if fila == n:
            soluciones.append(tablero[:])

        else:
            for columna in range(n):
                seguro = True
                # Comprobar si la nueva reina recibe ataque de alguna otra
                for i in range(fila):
                    # Comprobar la columna y las diagonales
                    if tablero[i] == columna or i - tablero[i] == fila - columna or i + tablero[i] == fila + columna:
                        seguro = False
                        break
                        
                # La posición es segura si no está atacada
                if seguro:
                    tablero[fila] = columna
                    resolver_n_reinas(tablero, fila + 1, soluciones)
def n_reinas(n):
    soluciones = []
    tablero = [-1] * n
    resolver_n_reinas(tablero, 0, soluciones)
    return soluciones
```

```python
n = 8
soluciones = n_reinas(n)

# Las soluciones vienen representadas por una lista de tamaño n en la que cada posición es 
# una fila del tablero y el valor es la columna en la que está colocada la reina
print(f"Número de posibles tableros de tamaño {n}x{n}: {len(soluciones)}")
for solucion in soluciones:
    print(solucion)
```
